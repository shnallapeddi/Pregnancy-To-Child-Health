```python
#Question 2
#Hypothesis 1

# Plot 1: Heatmap to show the correlation between maternal hypertension & mortality rates.
import seaborn as sns

plot.figure(figsize=(7, 5))
correlation_data = hypothesis_ds[['Hypertension to Birth Rate Ratio', 
                             'Infant Mortality Rate to Birth Rate Ratio',
                             'Neonatal Mortality Rate to Birth Rate Ratio']].corr()

sns.heatmap(correlation_data, annot=True, cmap='YlGnBu', linewidths=0.5)
plot.title('Correlation Heatmap of Hypertension & Mortality Rates')
plot.show()
```

The relationship between infant mortality rate to birth rate ratio, neonatal mortality rate to birth rate ratio, & hypertension to birth rate ratio is shown graphically in a heatmap. The range of correlation values is -1 to 1, where:

i. A complete positive correlation, or one in which both variables rise as one does, is denoted by a value of +1.
ii. A perfect negative correlation is represented by a value of -1, meaning that as one variable rises, the other falls.
iii. A value of 0 denotes no correlation.


a. Infant & neonatal mortality rates have a strong positive correlation (0.92):
According to this, areas with greater infant death rates also typically have higher neonatal mortality rates. It is anticipated that neonatal mortality deaths that occur during the first 28 days of life & infant mortality deaths that occur within the first year of life will be closely associated. Thus, it is likely that the same factors that impact neonatal mortality such as poor maternal health, early birth, & medical complications will also impact infant mortality as a whole.

b. Negative Correlation (-0.49) between Hypertension & Infant Mortality:
Infant mortality rate to birth rate ratio & hypertension to birth rate ratio have a moderately inverse association, as indicated by the correlation of -0.49, which is not significantly negative.

c. Negative Correlation (-0.37) between Hypertension & Neonatal Mortality:
This correlation, like the one with newborn mortality, indicates that there is a complex interaction between maternal hypertension & a moderate decrease in neonatal mortality. Although low birth weight & premature birth are known to be tied to maternal hypertension, the negative correlation observed here may indicate that improvements in medical treatment for hypertensive pregnancies are reducing some of these risks, which in some cases may lead to a lower neonatal mortality rate.

```python
#Question 2
#Hypothesis 2

#Second Plot(Different Kinds)
# Manually define thresholds for low/high categories
threshold_hospital_beds = hypothesis_ds['Hospital beds (per 1,000 people)'].median()
hypothesis_ds['Hospital Bed Category'] = np.where(hypothesis_ds['Hospital beds (per 1,000 people)'] > threshold_hospital_beds, 'High', 'Low')

threshold_health_staff = hypothesis_ds['Births attended by skilled health staff (% of total)'].median()
hypothesis_ds['Health Staff Category'] = np.where(hypothesis_ds['Births attended by skilled health staff (% of total)'] > threshold_health_staff, 'High', 'Low')

# Regression plot for hypertension vs infant mortality, grouped by hospital bed availability
plot.figure(figsize=(7, 6))
sns.lmplot(x='Prevalence of hypertension, pregnant women (% of pregnant women adults ages 30-79)', 
           y='Mortality rate, infant (per 1,000 live births)', 
           hue='Hospital Bed Category', data=hypothesis_ds, palette='coolwarm', height=6, aspect=1.2)
plot.title("Hypertension vs Infant Mortality: Impact of Hospital Bed Availability (Low/High)")
plot.grid(True)
plot.show()

# Regression plot for hypertension vs low birthweight, grouped by skilled health staff attendance 
plot.figure(figsize=(7, 6))
sns.lmplot(x='Prevalence of hypertension, pregnant women (% of pregnant women adults ages 30-79)', 
           y='Low-birthweight babies (% of births)', 
           hue='Health Staff Category', data=hypothesis_ds, palette='viridis', height=6, aspect=1.2)
plot.title("Hypertension vs Low Birthweight: Impact of Skilled Health Staff Attendance (Low/High)")
plot.grid(True)
plot.show()
```

Hypertension vs Infant Mortality (Grouped by Hospital Bed Availability):
Maternal hypertension has a less noticeable impact on infant mortality in areas with a high hospital bed availability. However, there is likely a higher positive correlation between newborn mortality & hypertension in areas with a shortage of hospital beds. This implies that the detrimental effects of maternal hypertension on infant mortality are mitigated in part by improved hospital facilities.

Hypertension vs Low Birthweight (Grouped by Skilled Health Staff Attendance):
Maternal hypertension might have less of an influence on low birthweight in areas with high attendance rates of trained healthcare workers. On the other hand, a higher proportion of low-birthweight infants may be more closely associated with maternal hypertension in areas with poor attendance of qualified medical professionals. This suggests that the negative effects of hypertension on birth outcomes are lessened when they have access to qualified healthcare professionals.
