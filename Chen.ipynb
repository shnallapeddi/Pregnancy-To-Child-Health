```python
#Data Cleaning
#Step 1 :  Remove duplicates rows

dataset.drop_duplicates(keep='first', inplace=True)

file_path = '/Users/sharanya/Downloads/Remove_Duplicate_Rows.csv' 
dataset.to_csv(file_path, index=False)
dataset.head()
```

```python
#Data Cleaning
#Step 2 : Remove irrelevant column like Year Code
print(dataset.columns)
dataset.columns = dataset.columns.str.strip()
dataset.drop(columns=['Year Code'], errors='ignore', inplace=True)
for col in dataset.columns:
    print(col)

file_path = '~/Downloads/remove_irrelevant_columns_dataset.csv'
dataset.to_csv(file_path, index=False)
```

```python
#Data Cleaning
#Step 3 : Remove irrelevant data

unwanted_regions = [ '', 'Fragile and conflict affected situations', 'Heavily indebted poor countries (HIPC)', 'High income', 'IBRD only', 'IDA & IBRD total', 'IDA blend', 'IDA only', 'IDA total', 'Late-demographic dividend', 'Early-demographic dividend', 'Post-demographic dividend', 'Pre-demographic dividend', 'Low & middle income', 'Least developed countries: UN classification', 'Low income', 'Lower middle income', 'Not classified', 'OECD members', 'Other small states', 'South Asia (IDA & IBRD)', 'Upper middle income', 'World' ]

dataset_cleaned = dataset[~dataset['Region Name'].isin(unwanted_regions)] 

file_path = '~/Downloads/remove_rows_unwanted_regions_dataset.csv'
dataset_cleaned.to_csv(file_path, index=False)

dataset_cleaned.head()
```

```python
print(dataset_cleaned.columns)
```

```python
#Data Cleaning
#Step 4 : Remove Missing Data

placeholder_value = ".."

# Calculate the threshold for dropping columns
threshold_80 = len(dataset_cleaned) * 0.8  # 80% threshold

# Remove columns where the count of placeholder values is greater than 80%
dataset_cleaned2 = dataset_cleaned.loc[:, dataset_cleaned.apply(lambda col: (col == placeholder_value).sum() <= threshold_80)]

# Display the cleaned DataFrame
print(dataset_cleaned2)

file_path = '~/Downloads/remove_missing_data.csv'
dataset_cleaned2.to_csv(file_path, index=False)

original_columns = set(dataset_cleaned.columns)

cleaned_columns = set(dataset_cleaned2.columns)

dropped_columns = original_columns - cleaned_columns

print("Dropped columns are listed below:", dropped_columns)
```

```python
replacements_tracker = pd.DataFrame(False, index=dataset_cleaned2.index, columns=dataset_cleaned2.columns)

# Data Cleaning 
#Step 5
for col in dataset_cleaned2.columns:
    # Identify cells with ".."
    mask = (dataset_cleaned2[col] == "..")
    total_count = len(dataset_cleaned2[col])
    missing_count = mask.sum()
    
    missing_percentage = missing_count / total_count

    if missing_percentage > 0.5:
        # Replace ".." with 0 for columns with >50% missing values
        dataset_cleaned2[col].replace("..", 0, inplace=True)
        replacements_tracker[col] = mask
    elif missing_percentage > 0:
        # Replace ".." with mean for columns with <=50% missing values
        dataset_cleaned2[col] = pd.to_numeric(dataset_cleaned2[col], errors='coerce')
        col_mean = dataset_cleaned2[col].mean(skipna=True)
        dataset_cleaned2[col].replace("..", col_mean, inplace=True)
        replacements_tracker[col] = mask

dataset_cleaned2.to_csv('cleaned_dataset_updated.csv', index=False)

replaced_rows = replacements_tracker.any(axis=1)
print("\nRows where '..' was replaced with a value:")
print(dataset_cleaned2[replaced_rows])
```

```python
#Data Cleaning
#Step 6 - Handling outliers

numeric_cols = dataset_cleaned2.select_dtypes(include=['float64', 'int64']).columns

def handle_outliers_iqr(dataset_cleaned2, col):
    Q1 = dataset_cleaned2[col].quantile(0.25)
    Q3 = dataset_cleaned2[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    median_value = dataset_cleaned2[col].median()

    # Replace outliers with the median value using .loc to avoid the warning
    dataset_cleaned2.loc[(dataset_cleaned2[col] < lower_bound) | (dataset_cleaned2[col] > upper_bound), col] = median_value

for col in numeric_cols:
    handle_outliers_iqr(dataset_cleaned2, col)

file_path = '~/Downloads/cleaned_dataset_outliers.csv'
dataset_cleaned2.to_csv(file_path, index=False)

# Display the first few rows of the cleaned dataset
dataset_cleaned2.head()
```

```python
 #Data cleaning
# Step 7 - Handling Negative Values
numeric_cols = dataset_cleaned2.select_dtypes(include=['float64', 'int64']).columns

dataset_cleaned2 = dataset_cleaned2[(dataset_cleaned2[numeric_cols] >= 0).all(axis=1)]

file_path = '~/Downloads/handling_negative_values.csv'
dataset_cleaned2.to_csv(file_path, index=False)

dataset_cleaned2.head()
```
